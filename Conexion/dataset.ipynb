```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("rohitgrewal/spotify-youtube-data")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
df = pd.read_csv('/kaggle/input/spotify-youtube-data/Spotify Youtube Dataset.csv')

# Mostrar las primeras filas del dataset
print("Primeras 5 filas del dataset:")
display(df.head())

# Mostrar información general del dataset
print("\nInformación general del dataset:")
display(df.info())

# Mostrar estadísticas descriptivas de las columnas numéricas
print("\nEstadísticas descriptivas:")
display(df.describe())

# Mostrar el número de valores únicos en cada columna
print("\nNúmero de valores únicos por columna:")
display(df.nunique())

# Verificar si hay valores nulos
print("\nValores nulos por columna:")
display(df.isnull().sum())

# Explorar la distribución de algunas columnas (ejemplo: 'Stream')
print("\nDistribución de las reproducciones (Stream):")
plt.figure(figsize=(10, 6))
sns.histplot(df['Stream'], bins=50, kde=True)
plt.title('Distribución de las Reproducciones (Stream)')
plt.xlabel('Reproducciones (Stream)')
plt.ylabel('Frecuencia')
plt.show()

# Explorar la relación entre algunas columnas (ejemplo: 'Loudness' y 'Energy')
print("\nRelación entre Loudness y Energy:")
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Loudness', y='Energy', data=df)
plt.title('Relación entre Loudness y Energy')
plt.xlabel('Loudness')
plt.ylabel('Energy')
plt.show()
```

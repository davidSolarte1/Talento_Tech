```python
#importar tablas
import pandas as pd
import numpy as np
import random


value_n_list = ['Andrew', 'Phillip', 'Jhon', 'David', 'Joe', 'Jose', 'Luis', 'Ana', 'Marta',  'Mario']
value_e_list = ['andrew@gmail.com', 'phillip@gmail.com', 'jhon@gmail.com', 'david@gmail.com', 'joe@gmail.com', 'jose@gmail.com', 'luis@gmail.com', 'ana@gmail.com', 'marta@gmail.com', 'mario@gmail.com']
value_w_list = [50, 78, 91.5, 50, 80, 95.5, 68, 68, 95, 59]
value_h_list = [1.75, 1.87, 1.69, 1.70, 1.90, 1.33, 1.55, 1.72, 1.59, 1.45]




data_dict = {"name":value_n_list, "weight":value_w_list, "height":value_h_list, "email":value_e_list}

for i in range(5):
    nombre = random.choice(value_n_list)
    peso = round(random.uniform(50, 100), 1)
    altura = round(random.uniform(1.40, 1.95), 2)
    correo = f"{nombre.lower()}@gmail.com"

    value_n_list.append(nombre)
    value_w_list.append(peso)
    value_h_list.append(altura)
    value_e_list.append(correo)

df=pd.DataFrame(data_dict)

df['bmi'] = df['weight'] / (df['height'] ** 2)
def clasificar_bmi(bmi):
    if bmi < 18.5:
        return "Bajo peso"
    elif bmi < 25:
        return "Normal"
    elif bmi < 30:
        return "Sobrepeso"
    elif bmi < 35:
        return "Obesidad I"
    elif bmi < 40:
        return "Obesidad II"
    else:
        return "Obesidad III"

# Aplicar clasificación al DataFrame
df['bmi_categoria'] = df['bmi'].apply(clasificar_bmi)

df['edad'] = np.random.randint(18, 90, size=len(df))

df
```
